#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here
employee.groupby("Department").agg({"Salary": np.mean}).reset_index().rename(columns={"Salary": "Mean-Salary"})
```

```python
employee_piv_dept = pd.pivot_table(employee, index=["Department"], values=["Salary"]).reset_index().rename(columns={"Salary": "Mean-Salary"})
```

```python
display(employee_piv_dept)
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
employee_piv_dept_tit = pd.pivot_table(employee, index=["Department", "Title"], values=["Salary"]).reset_index().rename(columns={"Salary": "Mean-Salary"})
```

```python
# your answer here
display(employee_piv_dept_tit)
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
# your answer here
pd.pivot_table(employee, index=["Title"], values=["Salary", "Years"], aggfunc=["sum"]).reset_index()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
pd.pivot_table(employee, index=["Department", "Title"], values=["Salary", "Name"], aggfunc={"Salary": np.mean, "Name": len}).reset_index().rename(columns={"Salary": "Mean-Salary", "Name": "Number of Employees"})
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
pivot_salary_data = pd.pivot_table(employee, index=["Department", "Title"], values=["Salary"], aggfunc=["median", "min", "max", "std" ]).reset_index()
```

```python
display(pivot_salary_data)
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
pd.pivot_table(employee, index=["Department", "Title"], values=["Years", "Salary"], aggfunc={"Years":sum, "Salary": np.mean}).reset_index().rename(columns={"Salary": "Mean-Salary"})
```
